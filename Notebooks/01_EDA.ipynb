```python
from sklearn.decomposition import PCA

X = df.drop("Class", axis=1)
y = df["Class"]

pca = PCA(n_components=2)
components = pca.fit_transform(X)

pca_df = pd.DataFrame(data=components, columns=["PC1","PC2"])
pca_df["Class"] = y
```

```python
plt.figure(figsize=(7,5))

normal = pca_df[pca_df["Class"]==0]
fraud = pca_df[pca_df["Class"]==1]

plt.scatter(normal["PC1"], normal["PC2"], alpha=0.3, label="Normal")
plt.scatter(fraud["PC1"], fraud["PC2"], alpha=0.9, label="Fraud")

plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.title("PCA Projection of Transactions")

plt.legend()
plt.show()
```

**Fraud detection is difficult because fraud overlaps with normal behaviour.**  
**Fraud appears closer to the boundary regions of the feature space, supporting the idea that fraudulent activity deviates from normal transaction behavior.**
